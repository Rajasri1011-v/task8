```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import pandas as pd

# Select features for PCA
features_for_pca = df[['Annual Income (k$)', 'Spending Score (1-100)']]

# Scale the features
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features_for_pca)

# Apply PCA
pca = PCA(n_components=2)
principal_components = pca.fit_transform(scaled_features)

# Create a new DataFrame for the principal components
df_pca = pd.DataFrame(data=principal_components, columns=['PC1', 'PC2'])

display(df_pca.head())
```

```python
from sklearn.preprocessing import StandardScaler

# Create a new DataFrame with 'Annual Income (k$)' and 'Spending Score (1-100)'
df_clustering = df[['Annual Income (k$)', 'Spending Score (1-100)']].copy()

# Initialize StandardScaler
scaler = StandardScaler()

# Fit and transform the data
df_clustering[['Annual Income (k$)', 'Spending Score (1-100)']] = scaler.fit_transform(df_clustering)

# Display the first few rows of the standardized data
display(df_clustering.head())
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Determine optimal k using the elbow method
inertia = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(df_clustering)
    inertia.append(kmeans.inertia_)

plt.figure(figsize=(8, 6))
plt.plot(range(1, 11), inertia, marker='o')
plt.title('Elbow Method for Optimal k')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.show()

# Based on the elbow method, choose k = 5 (this can be adjusted based on the plot)
optimal_k = 5

# Initialize and fit the KMeans model
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
kmeans.fit(df_clustering)

# Add cluster labels to the DataFrame
df_clustering['cluster_label'] = kmeans.labels_

# Display the first few rows of the DataFrame with cluster labels
display(df_clustering.head())
```

```python
import matplotlib.pyplot as plt

# Create a scatter plot of the data points, colored by cluster
plt.figure(figsize=(10, 6))  # Adjust figure size for better visualization
for cluster_id in df_clustering['cluster_label'].unique():
    cluster_data = df_clustering[df_clustering['cluster_label'] == cluster_id]
    plt.scatter(cluster_data['Annual Income (k$)'], cluster_data['Spending Score (1-100)'], label=f'Cluster {cluster_id}', alpha=0.7)

# Plot the centroids
centroids = kmeans.cluster_centers_
plt.scatter(centroids[:, 0], centroids[:, 1], s=100, c='black', marker='X', label='Centroids')

# Customize the plot
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.title('Customer Segmentation based on Annual Income and Spending Score')
plt.legend(title='Clusters', loc='upper right')
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler

# Create df_clustering
df_clustering = df[['Annual Income (k$)', 'Spending Score (1-100)']].copy()

# Scale the data
scaler = StandardScaler()
df_clustering = pd.DataFrame(scaler.fit_transform(df_clustering), columns = df_clustering.columns)

display(df_clustering.head())
```

```python
from sklearn.cluster import KMeans

kmeans_models = {}
for n_clusters in range(2, 11):  # Experiment with 2 to 10 clusters
    kmeans = KMeans(n_clusters=n_clusters, random_state=0, n_init=10)
    kmeans.fit(df_clustering)
    kmeans_models[n_clusters] = kmeans
```

```python
from sklearn.metrics import silhouette_score

silhouette_scores = {}
for n_clusters, kmeans in kmeans_models.items():
    cluster_labels = kmeans.labels_
    score = silhouette_score(df_clustering, cluster_labels)
    silhouette_scores[n_clusters] = score
print(silhouette_scores)
```

```python
import matplotlib.pyplot as plt

# Find the optimal number of clusters based on the highest Silhouette Score
optimal_n_clusters = max(silhouette_scores, key=silhouette_scores.get)
optimal_score = silhouette_scores[optimal_n_clusters]

# Plot the Silhouette Scores
plt.figure(figsize=(10, 6))
plt.plot(silhouette_scores.keys(), silhouette_scores.values(), marker='o')
plt.xlabel("Number of Clusters")
plt.ylabel("Silhouette Score")
plt.title("Silhouette Score for Different Numbers of Clusters")
plt.annotate(f"Optimal Clusters: {optimal_n_clusters} (Score: {optimal_score:.4f})",
             xy=(optimal_n_clusters, optimal_score),
             xytext=(optimal_n_clusters + 0.5, optimal_score + 0.05),  # Adjust text position
             arrowprops=dict(facecolor='black', shrink=0.05))
plt.show()
```
